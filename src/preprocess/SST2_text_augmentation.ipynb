````python
!textattack augment --input-csv sst2_train.csv --output-csv sst2_train_eda.csv  --input-column text --recipe eda --pct-words-to-swap .1 --transformations-per-example 1 --exclude-original
````

The output CSV file might be malformed, so we fix it manually.

````python
def fix_csv_line(line: str) -> list[str]:
  tokens = line.strip().rsplit(',', 1)
  tokens[0] = tokens[0].strip('"').strip()
  return tokens

def fix_csv_file(path: str):
  with open(path, 'r') as f:
    lines = f.readlines()
    lines = lines[1:] # skip labels
    lines = [fix_csv_line(line) for line in lines]
    df = pd.DataFrame(lines, columns=['text', 'label'])
    df.to_csv("fixed_" + path, index=False)

fix_csv_file('sst2_train_eda.csv')
````
